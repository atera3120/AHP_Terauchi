# 卒業研究 提案手法まとめ  

***
## ALD法

人間の感覚は非線形で，おおむね対数関数で近似されることが知られている．そのため，AHPにおいてもこの非線形性を考慮する．一対比較行列から重要度のクリスプな値を推定するAHPにおいて，以下のような対数空間内での最適化を考える．対数空間内の最適化を用いることから，外れ値の影響を受けにくくなると予想される．

$$
\begin{align*}
\operatorname{minimize} ~~ & ~~
\sum_{i \in N} \sum_{j < i} \left| \ln \left( a_{ij} \right) - \ln \left( \frac{w_i}{w_j} \right) \right| \\
\end{align*}
$$

上式は非線形計画問題であるが，以下のように文字を置き換えることによって，線形計画問題に帰着する．

$$
\begin{align*}
\operatorname{minimize} ~~ & ~~
\sum_{i \in N} \sum_{j < i} U_{ij}, \\
\text{subject to} ~~ & ~~
\ln \left( a_{ij} \right) - u_i + u_j \leq U_{ij}, &&
i \in N, ~~ & j<i \\ & ~~
u_i - u_j -\ln \left( a_{ij} \right) \leq U_{ij}, &&
i \in N, ~~ & j<i \\ & ~~
u_i \in \mathbb{R}.
\end{align*}
$$


$$where ~~ u_i = \ln \left( w_i \right), ~ U_{ij} = \left| \ln \left( a_{ij} \right) - u_i + u_j \right|, ~ i \in N$$

ここで，求まった最適解から
$$w_i = \lambda \exp \left( u_i \right)$$
として重みを計算することができるが，この際，総和が１になるように定める．

## MMR_E, MMR_G, MMR_A

**Phase 1**  
区間重要度の中心値推定による弱モデル 
* $k$ 番目を除いた $n-1$ 個の $w_i$ に対して EV, GM, ALD で $w_{i}^{C}(k)$ を求める（総和は１）
* 上記で求めた $w_{i}^{C}(k)$ の和を $\mu_k$ 倍することで， $k$ 番目の区間重要度中心は $1-\mu_k$ となる．　　

上記の操作によって，$k$番目の重みに自由度を残しつつ，その他の重みの比を固定することができる．  
このような弱モデルが$N$個生成される．

**Phase 2**  
各 $k$ に対して以下の線形最適化問題を解き，最適値を $d_{\overline{k}}^{*}$ とする：  
$$
\begin{align*}

\operatorname{minimize} ~~ & ~~
d_{\overline{k}} = \sum_{j \in N \backslash k} l_j, \\

\text{subject to} ~~ & ~~
a_{ij} \left( \mu_k w_{j}^{C} - l_j \right) \leq \mu_k w_{i}^{C} + l_i, &&
i \neq j, k, ~~ & j \neq k \\ & ~~

a_{kj} \left( \mu_k w_{j}^{C} - l_j \right) \leq 1-\mu_k + l_k, &&
j \neq k \\ & ~~

a_{ik} \left\{ \left(1- \mu_k \right) - l_k \right\} \leq \mu_k w_{i}^{C} + l_i, &&
i \neq k \\ & ~~

\left( 1 - \mu_k \right) + l_k + \sum_{i \in N \backslash \left\{ j, k \right\} } \left( \mu_k w_{i}^{C} + l_i \right) + \mu_k w_{j}^{C} - l_j \geq 1, &&
i \in N \backslash k \\ & ~~

\left( 1 - \mu_k \right) - l_k + \sum_{i \in N \backslash \left\{ j, k \right\}} \left( \mu_k w_{i}^{C} - l_i \right) + \mu_k w_{j}^{C} + l_j \leq 1, &&
i \in N \backslash k \\ & ~~


\sum_{i \in N \backslash k} \left( \mu_k w_{i}^{C} + l_i \right) + \left( 1 - \mu_k \right) - l_k \geq 1 \\ & ~~
\sum_{i \in N \backslash k} \left( \mu_k w_{i}^{C} - l_i \right) + \left( 1 - \mu_k \right) + l_k \leq 1 \\ & ~~

\mu_k w_{i}^{C} - l_i \geq \epsilon, &&
i \in N \backslash k ~~ & \\ & ~~

\left( 1 - \mu_k \right) - l_k \geq \epsilon \\ & ~~


\end{align*}
$$

<div style="page-break-before:always"></div>


**Phase 3**  
各 $k$ に対して以下の線形最適化問題を解き，最適値を $l_{k}^{*}$ とする：  
$$
\begin{align*}
\operatorname{minimize} ~~ & ~~
l_k \\
\text{subject to} ~~ & ~~
a_{ij} \left( \mu_k w_{j}^{C} - l_j \right) \leq \mu_k w_{i}^{C} + l_i, &&
i \neq j, k, ~~ & j \neq k \\ & ~~

a_{kj} \left( \mu_k w_{j}^{C} - l_j \right) \leq 1-\mu_k + l_k, &&
j \neq k \\ & ~~

a_{ik} \left\{ \left(1- \mu_k \right) - l_k \right\} \leq \mu_k w_{i}^{C} + l_i, &&
i \neq k \\ & ~~

\left( 1 - \mu_k \right) + l_k + \sum_{i \in N \backslash \left\{ j, k \right\} } \left( \mu_k w_{i}^{C} + l_i \right) + \mu w_{j}^{C} - l_j \geq 1, &&
i \in N \backslash k \\ & ~~

\left( 1 - \mu_k \right) - l_k + \sum_{i \in N \backslash \left\{ j, k \right\} } \left( \mu_k w_{i}^{C} - l_i \right) + \mu_k w_{j}^{C} + l_j \leq 1, &&
i \in N \backslash k \\ & ~~

\sum_{i \in N \backslash k} \left( \mu_k w_{i}^{C} + l_i \right) + \left( 1 - \mu_k \right) - l_k \geq 1 \\ & ~~

\sum_{i \in N \backslash k} \left( \mu_k w_{i}^{C} - l_i \right) + \left( 1 - \mu_k \right) + l_k \leq 1 \\ & ~~

\sum_{j \in N \backslash k} l_j = d_{\overline{k}}^{*}\\ & ~~

\mu_k w_{i}^{C} - l_i \geq \epsilon, &&
i \in N \backslash k ~~ & \\ & ~~

\left( 1 - \mu_k \right) - l_k \geq \epsilon \\ & ~~


\end{align*}
$$

**Phase 4**  
Phase 3 で導かれた結果は$N$個の弱モデル分だけ考えられるため．以下のように結果を統合する：  
* 各$l_{k}^{*}$に対して，  
$$\left\{\begin{array}{l}w_i(k)=\left[\mu_{k}^* w_i^c-l_{k}^{*},\:\: \mu_{k}^* w_i^c+l_{k}^{*}\right], \quad i \neq k \\ w_i(k)=\left[\left(1-\mu_{k}^*\right)-l_{k}^{*},\:\:\left(1-\mu_{k}^*\right)+l_{k}^{*}\right], \quad i=k\end{array}\right.$$
  

$$\overline{w}_i^L=\min _{k \in N} w_i^L(k) \quad \overline{w}_i^U=\max _{k \in N} w_i^U(k)$$  
  
ここで，$\overline{w}_i^L$ と $\overline{w}_i^U$ の中心 $\overline{w}_i^C$の総和が1となるように正規化する：
  
$$\overline{\overline{w}}_i^L=\frac{\overline{w}_i^L}{\sum_{i \in N} \frac{\overline{w}_i^L+\overline{w}_i^U}{2}} \quad \overline{\overline{w}}_i^U=\frac{\overline{w}_i^U}{\sum_{i \in N} \frac{\overline{w}_i^L+\overline{w}_i^U}{2}}$$

## MMR_E*, MMR_G*, MMR_A*

MMR-analysisの結果から明らかなように，MMRでは最適化の過程である特定の位置に「重み」が集中し，それ以外の場所は極端に小さな重みを取るというアンバランスが発生した．
この状況を改善するために，以下のように「区間重要度幅に対する重み（比率）」の最小化を図る　　

**Phase1, 4** は上記MMRと同様にして，**Phase2, 3** の最適化部分を以下のように変更する

**Phase 2【非線形】**  
各 $k$ に対して、以下の最適化問題を解き、最適値を $d_{\overline{k}}^{*}$ とする：  
（目的関数を重みと幅の比で扱うために，これまでの各式に対して，$\dfrac{1}{\mu_k}$ 倍している）  
$$
\begin{align*}
\operatorname{minimize} ~~ & ~~

d_{\overline{k}} = \sum_{j \in N \backslash k} \frac{l_j}{\mu_k}, \\
\text{subject to} ~~ & ~~

a_{ij} \left( w_{j}^{C} - \frac{l_j}{\mu_k} \right) \leq w_{i}^{C} + \frac{l_i}{\mu_k}, &&
i \neq j, k, ~~ & j \neq k \\ & ~~

a_{kj} \left( w_{j}^{C} - \frac{l_j}{\mu_k} \right) \leq \frac{1-\mu_k}{\mu_k} + \frac{l_k}{\mu_k}, &&
j \neq k \\ & ~~

a_{ik} \left(\frac{1-\mu_k}{\mu_k} - \frac{l_k}{\mu_k} \right) \leq w_{i}^{C} + \frac{l_i}{\mu_k}, &&
i \neq k \\ & ~~

\frac{1-\mu_k}{\mu_k} + \frac{l_k}{\mu_k} + \sum_{i \in N \backslash \left\{ j, k \right\} } \left( w_{i}^{C} + \frac{l_i}{\mu_k} \right) + w_{j}^{C} - \frac{l_j}{\mu_k} \geq \frac{1}{\mu_k}, &&
i \in N \backslash k \\ & ~~

\frac{1-\mu_k}{\mu_k} - \frac{l_k}{\mu_k} + \sum_{i \in N \backslash \left\{ j, k \right\} } \left( w_{i}^{C} - \frac{l_i}{\mu_k} \right) + w_{j}^{C} + \frac{l_j}{\mu_k} \leq \frac{1}{\mu_k}, &&
i \in N \backslash k \\ & ~~

\sum_{i \in N \backslash k} \left( w_{i}^{C} + \frac{l_i}{\mu_k} \right) + \frac{1-\mu_k}{\mu_k} - \frac{l_k}{\mu_k} \geq \frac{1}{\mu_k} \\ & ~~

\sum_{i \in N \backslash k} \left( w_{i}^{C} - \frac{l_i}{\mu_k} \right) + \frac{1-\mu_k}{\mu_k} + \frac{l_k}{\mu_k} \leq \frac{1}{\mu_k} \\ & ~~

w_{i}^{C} - \frac{l_i}{\mu_k} \geq \frac{\epsilon}{\mu_k}, &&
i \in N, ~~ & i \neq k \\ & ~~

\frac{1-\mu_k}{\mu_k} - \frac{l_k}{\mu_k} \geq \frac{\epsilon}{\mu_k} \\ & ~~


\end{align*}
$$

<div style="page-break-before:always"></div>

ただし，この問題は非線形最適化問題となるため，  
$\dfrac{l_i}{\mu_k} = L_{ki}, ~~ \dfrac{1}{\mu_k} = t_k$  
と置き換えることで線形最適化問題に持ち込む：

**Phase 2' 【線形】**  
各 $k$ に対して、以下の線形最適化問題を解き、最適値を $d_{\overline{k}}^{*}$ とする：  
$$
\begin{align*}
\operatorname{minimize} ~~ & ~~

d_{\overline{k}} = \sum_{j \in N \backslash k} L_{kj}, \\
\text{subject to} ~~ & ~~

a_{ij} \left( w_{j}^{C} - L_{kj} \right) \leq w_{i}^{C} + L_{ki}, &&
i \neq j, k, ~~ & j \neq k \\ & ~~

a_{kj} \left( w_{j}^{C} - L_{kj} \right) \leq t_k-1 + L_{kk}, &&
j \neq k \\ & ~~

a_{ik} \left\{ \left( t_k -1 \right) - L_{kk} \right\} \leq w_{i}^{C} + L_{ki}, &&
i \neq k \\ & ~~

\left( t_k-1 \right) + L_{kk} + \sum_{i \in N \backslash \left\{ j, k \right\} } \left( w_{i}^{C} + L_{ki} \right) + w_{j}^{C} - L_{kj} \geq t_k, &&
i \in N \backslash k \\ & ~~

\left( t_k-1 \right) - L_{kk} + \sum_{i \in N \backslash \left\{ j, k \right\} } \left( w_{i}^{C} - L_{ki} \right) + w_{j}^{C} + L_{kj} \leq t_k, &&
i \in N \backslash k \\ & ~~

\sum_{i \in N \backslash k} \left( w_{i}^{C} + L_{ki} \right) + \left( t_k-1 \right) - L_{kk} \geq t_k \\ & ~~

\sum_{i \in N \backslash k} \left( w_{i}^{C} - L_{ki} \right) + \left( t_k-1 \right) + L_{kk} \leq t_k \\ & ~~

w_{i}^{C} - L_{ki} \geq \epsilon , &&
i \in N, ~~ & i \neq k \\ & ~~

\left( t_k-1 \right) - L_{kk} \geq \epsilon \\ & ~~


\end{align*}
$$

<div style="page-break-before:always"></div>


**Phase 3【非線形】**  
各 $k$ に対して、以下の線形最適化問題を解き、最適値を $l_{k}^{*}$ とする：  
（目的関数を重みと幅の比で扱うために，これまでの各式に対して，$\dfrac{1}{1-\mu_k}$ 倍している）  
$$
\begin{align*}
\operatorname{minimize} ~~ & ~~
\frac{l_k}{1-\mu_k} \\
\text{subject to} ~~ & ~~

a_{ij} \left( \frac{\mu_k}{1-\mu_k} w_{j}^{C} - \frac{l_j}{1-\mu_k} \right) \leq \frac{\mu_k}{1-\mu_k} w_{i}^{C} + \frac{l_i}{1-\mu_k}, &&
i \neq j, k, ~~ & j \neq k \\ & ~~

a_{kj} \left( \frac{\mu_k}{1-\mu_k} w_{j}^{C} - \frac{l_j}{1-\mu_k} \right) \leq 1 + \frac{l_k}{1-\mu_k}, &&
j \neq k \\ & ~~

a_{ik} \left( 1 - l_k \right) \leq \frac{\mu_k}{1-\mu_k} w_{i}^{C} + \frac{l_i}{1-\mu_k}, &&
i \neq k \\ & ~~

1 + \frac{l_k}{1-\mu_k} + \sum_{i \in N \backslash \left\{ j, k \right\} } \left( \frac{\mu_k}{1-\mu_k} w_{i}^{C} + \frac{l_i}{1-\mu_k} \right) + \frac{\mu_k}{1-\mu_k} w_{j}^{C} - \frac{l_j}{1-\mu_k} \geq \frac{1}{1-\mu_k}, &&
i \in N \backslash k \\ & ~~

1 - \frac{l_k}{1-\mu_k} + \sum_{i \in N \backslash \left\{ j, k \right\} } \left( \frac{\mu_k}{1-\mu_k} w_{i}^{C} - \frac{l_i}{1-\mu_k} \right) + \frac{\mu_k}{1-\mu_k} w_{j}^{C} + \frac{l_j}{1-\mu_k} \leq \frac{1}{1-\mu_k}, &&
i \in N \backslash k \\ & ~~

\sum_{i \in N \backslash k} \left( \frac{\mu_k}{1-\mu_k} w_{i}^{C} + \frac{l_i}{1-\mu_k} \right) + 1 - \frac{l_k}{1-\mu_k} \geq \frac{1}{1-\mu_k} \\ & ~~

\sum_{i \in N \backslash k} \left( \frac{\mu_k}{1-\mu_k} w_{i}^{C} - \frac{l_i}{1-\mu_k} \right) + 1 + \frac{l_k}{1-\mu_k} \leq \frac{1}{1-\mu_k} \\ & ~~

\sum_{j \in N \backslash k} \frac{l_j}{1-\mu_k} = \frac{\mu_k}{1-\mu_k} d_{\overline{k}}^{*}\\ & ~~

\frac{\mu_k}{1-\mu_k} w_{i}^{C} - \frac{l_i}{1-\mu_k} \geq \epsilon, &&
i \in N \backslash k ~~ & \\ & ~~

1 - \frac{l_k}{1-\mu_k} \geq \epsilon \\ & ~~


\end{align*}
$$

<div style="page-break-before:always"></div>

ただし，この問題は非線形最適化問題となるため，  
$\dfrac{l_i}{1-\mu_k} = L'_{ki}, ~~ \dfrac{1}{1-\mu_k} = t_k'$  
と置き換えることで線形最適化問題に持ち込む：

**Phase 3'【線形】**  
各 $k$ に対して、以下の線形最適化問題を解き、最適値を $l_{k}^{*}$ とする：  
$$
\begin{align*}
\operatorname{minimize} ~~ & ~~
L'_{kk} \\
\text{subject to} ~~ & ~~
a_{ij} \left\{ \left( t_k'-1 \right) w_{j}^{C} - L'_{kj} \right\} \leq \left( t_k'-1 \right) w_{i}^{C} + L'_{ki}, &&
i \neq j, k, ~~ & j \neq k \\ & ~~

a_{kj} \left\{ \left( t_k'-1 \right) w_{j}^{C} - L'_{kj} \right\} \leq 1 + L'_{kk}, &&
j \neq k \\ & ~~

a_{ik} \left(1 - L'_{kk} \right) \leq \left( t_k'-1 \right) w_{i}^{C} + L'_{ki}, &&
i \neq k \\ & ~~

1 + L'_{kk} + \sum_{i \in N \backslash \left\{ j, k \right\} } \left\{ \left( t_k'-1 \right) w_{i}^{C} + L'_{ki} \right\} + \left( t_k'-1 \right) w_{j}^{C} - L'_{kj} \geq t_k', &&
i \in N \backslash k \\ & ~~

1 - L'_{kk} + \sum_{i \in N \backslash \left\{ j, k \right\} } \left\{ \left( t_k'-1 \right) w_{i}^{C} - L'_{ki} \right\} + \left( t_k'-1 \right) w_{j}^{C} + L'_{kj} \leq t_k', &&
i \in N \backslash k \\ & ~~

\sum_{i \in N \backslash k} \left\{ \left( t_k'-1 \right) w_{i}^{C} + L'_{ki} \right\} + 1 - L'_{kk} \geq t_k' \\ & ~~

\sum_{i \in N \backslash k} \left\{ \left( t_k'-1 \right) w_{i}^{C} - L'_{ki} \right\} + 1 + L'_{kk} \leq t_k' \\ & ~~

\sum_{j \in N \backslash k} L'_{kj} = \left( t_k'-1 \right) d_{\overline{k}}^{*}\\ & ~~

\left( t_k'-1 \right) w_{i}^{C} - L'_{ki} \geq \epsilon, &&
i \in N \backslash k ~~ & \\ & ~~

1 - L'_{kk} \geq \epsilon \\ & ~~


\end{align*}

## AMR_E. AMR_G, AMR_A

MMR法の**Phase 4**を以下のように変更する：  
**Phase 4**  
Phase 3 で導かれた結果は$N$個の弱モデル分だけ考えられるため．以下のように結果を統合する：   
* 各$l_{k}^{*}$に対して、  
$$\left\{\begin{array}{l}w_i(k)=\left[\mu_{k}^* w_i^c-l_{k}^{*},\:\: \mu_{k}^* w_i^c+l_{k}^{*}\right], \quad i \neq k \\ w_i(k)=\left[\left(1-\mu_{k}^*\right)-l_{k}^{*},\:\:\left(1-\mu_{k}^*\right)+l_{k}^{*}\right], \quad i=k\end{array}\right.$$
  

$$\overline{w}_i^L=\frac{1}{n} \sum_{i \in N} w_i^L(k) \quad \overline{w}_i^U=\frac{1}{n} \sum_{i \in N} w_i^U(k)$$  

この結果は，正規性を満たしている．
